```python
#dopisać jeszcze opcje jeżeli gracz wpisze liczbę spoza zakresu 1-9

plansza = ["-", "-", "-", "-", "-", "-", "-", "-", "-"]


print("Kółko i krzyżyk!")
def pokaż_plansze():
    print(plansza[0] + plansza[1] + plansza[2])
    print(plansza[3] + plansza[4] + plansza[5])
    print(plansza[6] + plansza[7] + plansza[8])

def gra():
    pokaż_plansze()
    
    while wygrana() is False:
        if plansza.count("-") == 0:
            print("Koniec gry. Brak wolnych miejsc!")
            break            
        ruch_x()
        wygrana()
        if wygrana() is True:
            print("Wygrał X")
            break
        if plansza.count("-") == 0:
            print("Koniec gry. Brak wolnych miejsc!")
            break 
        ruch_o()
        wygrana()
        if wygrana() is True:
            print("Wygrał O")
            break 

def ruch_x():
    
    ruch_x = input("Gdzie postawić 'x':")
    print(f"Gdzie postawić 'x': {ruch_x}")
    ruch_x = int(ruch_x) - 1
    if plansza[ruch_x] == "-":
        plansza[ruch_x] = "x"
        pokaż_plansze()
    else:
        print("Nie można nadpisać zaznaczonego pola! Spróbuj jeszcze raz!")
        ruch_x = input("Gdzie postawić 'x':")
        print(f"Gdzie postawić 'x': {ruch_x}")
        ruch_x = int(ruch_x) - 1
        plansza[ruch_x] = "x"
        pokaż_plansze()

def ruch_o():
    ruch_o = input("Gdzie postawić 'o':")
    print(f"Gdzie postawić 'o': {ruch_o}")
    ruch_o = int(ruch_o) - 1
    if plansza[ruch_o] == "-":
        plansza[ruch_o] = "o"
        pokaż_plansze()
    else:
        print("Nie można nadpisać zanaczonego pola! Spróbuj jeszcze raz!")
        ruch_o = input("Gdzie postawić 'o':")
        print(f"Gdzie postawić 'x': {ruch_o}")
        ruch_o = int(ruch_o) - 1
        plansza[ruch_o] = "o"
        pokaż_plansze()
        

def wygrana():
    if plansza[0] == plansza[1] == plansza[2] == "x":
        return True
    elif plansza[3] == plansza[4] == plansza[5] == "x":
        return True
    elif plansza[6] == plansza[7] == plansza[8] == "x":
        return True
    elif plansza[0] == plansza[3] == plansza[6] == "x":
        return True    
    elif plansza[1] == plansza[4] == plansza[7] == "x":
        return True
    elif plansza[2] == plansza[5] == plansza[8] == "x":
        return True   
    elif plansza[0] == plansza[4] == plansza[8] == "x":
        return True
    elif plansza[2] == plansza[4] == plansza[6] == "x":
        return True
    elif plansza[0] == plansza[1] == plansza[2] == "o":
        return True
    elif plansza[3] == plansza[4] == plansza[5] == "o":
        return True
    elif plansza[6] == plansza[7] == plansza[8] == "o":
        return True
    elif plansza[0] == plansza[3] == plansza[6] == "o":
        return True    
    elif plansza[1] == plansza[4] == plansza[7] == "o":
        return True
    elif plansza[2] == plansza[5] == plansza[8] == "o":
        return True   
    elif plansza[0] == plansza[4] == plansza[8] == "o":
        return True
    elif plansza[2] == plansza[4] == plansza[6] == "o":
        return True
    else:
        return False

gra()
```
